### **use function in your table**

`pd.read_excel(converters = {column you want to use your function in: function} )`

```python
def convert_people_cell(cell):
    if cell == 'n.a.':
        return 'sam waltuh'
    return cell

def convert_eps_cell(cell):
    if cell == 'not available':
        return None
    return cell
```

```python
df_4 = pd.read_excel('C:/Users/Bayram/Downloads/stock_data.xlsx', 'Sheet1', converters={
                    'people': convert_people_cell,
    'eps': convert_eps_cell
})
df_4
```

**you can do it while importing your table, to avoid messy data early**

### you have 2 dataframes, you want to export them in excel file format to different sheets

`with pd.ExcelWriter('name of file') as writer:`

    `dataframe1.to_excel(writer, sheet_name='name of choice')`

**

# Merge Dataframes

```python
df_9_0 = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Orlando'],
    'temperature': [21,14,35]
})
df_9_0
```

```python
df_9_1 = pd.DataFrame({
    'city': ['Chicago', 'New York', 'Orlando'],
    'humidity': [65,68,75]
})
df_9_1
```

```python
merged = pd.merge(df_9_0, df_9_1, on='city')
merged
```

```python
df_9_2 = pd.DataFrame({
    'city': ['New York', 'Chicago', 'Orlando', 'Baltimore'],
    'temperature': [21,14,35,32]
})

df_9_3 = pd.DataFrame({
    'city': ['Chicago', 'New York', 'San Francisco'],
    'humidity': [65,68,71]
})

print(df_9_2, '\n'*3, df_9_3)
```

```python
merged_0 = pd.merge(df_9_2, df_9_3, on='city', how='outer')
merged_0
```

```python
merged_1 = pd.merge(df_9_2, df_9_3, on='city', how='left')
merged_1
```

```python
merged_2 = pd.merge(df_9_2, df_9_3, on='city', how='right')
merged_2
```

```python
merged_0 = pd.merge(df_9_2, df_9_3, on='city', how='outer', indicator=True)
merged_0
```

```python
df_9_4 = pd.DataFrame({
    'city':['New York', 'Chicago', 'Orlando', 'Baltimore'],
    'temperature':[21, 14, 35, 38],
    'humidity':[65, 68, 71, 75]
})
df_9_4
```

```python
df_9_5 = pd.DataFrame({
    'city':['Chicago', 'New York', 'San Diego'],
    'temperature':[21, 14, 35],
    'humidity':[65, 68, 71]
})
df_9_5
```

```python
merged_3 = pd.merge(df_9_4, df_9_5, on='city')
merged_3
```

```python
merged_4 = pd.merge(df_9_4, df_9_5, on='city', suffixes=('_left', '_right'))
merged_4
```
